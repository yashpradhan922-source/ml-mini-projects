```python
from transformers import pipeline

# Load sentiment analysis pipeline
classifier = pipeline("sentiment-analysis")

# List of sentences to analyze
sentences = [
    "I love this AI course",
    "This course is very difficult",
    "The instructor explains concepts very well"
]

# Run sentiment analysis
results = classifier(sentences)

# Print results clearly
for sentence, result in zip(sentences, results):
    print("Sentence:", sentence)
    print("Sentiment:", result['label'])
    print("Confidence:", round(result['score'], 3))
    print("-" * 40)
```

```python
from transformers import pipeline
from pypdf import PdfReader
import re # Import regex for splitting text into sentences

# Load sentiment analysis pipeline
classifier = pipeline("sentiment-analysis")

# Path to your PDF file
pdf_path = '/content/pythonBOOK (1).pdf'

# Extract text from the PDF
def extract_text_from_pdf(pdf_path):
    reader = PdfReader(pdf_path)
    text = ""
    for page in reader.pages:
        text += page.extract_text() + "\n"
    return text

extracted_text = extract_text_from_pdf(pdf_path)

# Split the extracted text into sentences (a simple approach)
sentences = re.split(r'(?<=[.!?]) +', extracted_text)

# Filter out any empty strings that might result from splitting
sentences = [s.strip() for s in sentences if s.strip()]

# Take only the first few sentences if the document is very long
# to avoid excessive processing time for a demonstration
max_sentences = 10
if len(sentences) > max_sentences:
    print(f"PDF contains {len(sentences)} sentences. Analyzing the first {max_sentences} for demonstration.")
    sentences = sentences[:max_sentences]

if not sentences:
    print("No text extracted or no sentences found in the PDF.")
else:
    # Run sentiment analysis with truncation for long sentences
    # The default max_length for the model is 512, so we'll explicitly set it for clarity.
    results = classifier(sentences, max_length=512, truncation=True)

    # Print results clearly
    for sentence, result in zip(sentences, results):
        print("Sentence:", sentence)
        print("Sentiment:", result['label'])
        print("Confidencezele:", round(result['score'], 3))
        print("-" * 40)
```

```python
import nltk
nltk.download('vader_lexicon')

from nltk.sentiment import SentimentIntensityAnalyzer

# create analyzer
sia = SentimentIntensityAnalyzer()

# input text
text = "This laptop is very good but battery life is bad"

# get sentiment score
score = sia.polarity_scores(text)

print("Sentiment Scores:", score)

# classification
compound = score['compound']

if compound >= 0.05:
    print("Sentiment: Positive")
elif compound <= -0.05:
    print("Sentiment: Negative")
else:
    print("Sentiment: Neutral")
```

```python
import random
import nltk
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB

nltk.download('punkt')
nltk.download('wordnet')

lemmatizer = WordNetLemmatizer()

# Training Data
training_sentences = [
    "hello",
    "hi",
    "how are you",
    "what is your name",
    "bye",
    "goodbye",
    "thanks",
    "thank you"
]

training_labels = [
    "greeting",
    "greeting",
    "status",
    "name",
    "bye",
    "bye",
    "thanks",
    "thanks"
]

# Responses
responses = {
    "greeting": ["Hello!", "Hi there!", "Hey!"],
    "status": ["I am fine!", "Doing great!", "I'm good!"],
    "name": ["I am an NLP chatbot.", "My name is Python Bot."],
    "bye": ["Goodbye!", "See you later!"],
    "thanks": ["You're welcome!", "No problem!"]
}

# NLP preprocessing
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(training_sentences)

# ML Model
model = MultinomialNB()
model.fit(X, training_labels)

print("Chatbot is ready! Type 'exit' to stop.")

while True:
    user_input = input("You: ")

    if user_input.lower() == "exit":
        print("Bot: Goodbye!")
        break

    input_vector = vectorizer.transform([user_input])
    prediction = model.predict(input_vector)[0]

    response = random.choice(responses[prediction])
    print("Bot:", response)
```

```python
import nltk
import random
from nltk.util import ngrams
from collections import defaultdict

nltk.download('punkt')

# Training text dataset
text = """
machine learning is very interesting
machine learning helps computers learn from data
data science uses machine learning algorithms
python is popular for machine learning projects
artificial intelligence and machine learning are related fields
"""

# Tokenization
tokens = nltk.word_tokenize(text.lower())

# Create bigram model
bigrams = list(ngrams(tokens, 2))

model = defaultdict(list)

for w1, w2 in bigrams:
    model[w1].append(w2)

# Function to generate sentence
def generate_sentence(start_word, length=10):

    word = start_word.lower()
    sentence = [word]

    for i in range(length):
        if word not in model:
            break

        next_word = random.choice(model[word])
        sentence.append(next_word)
        word = next_word

        if next_word == ".":
            break

    return " ".join(sentence)

# Chat style input
while True:

    user_word = input("\nEnter a word to generate sentence (type exit to stop): ")

    if user_word.lower() == "exit":
        print("Program stopped")
        break

    if user_word.lower() not in model:
        print("Word not in training data. Try another word.")
    else:
        generated = generate_sentence(user_word)
        print("Generated Sentence:", generated)
```

```python
from transformers import pipeline

# Load text generation model
generator = pipeline("text-generation", model="gpt2")

print("---- NLP Text Generator ----")

while True:
    user_input = input("\nEnter a starting sentence (or type 'exit'): ")

    if user_input.lower() == "exit":
        break

    # Generate text
    result = generator(
        user_input,
        max_length=80,
        num_return_sequences=1,
        temperature=0.7
    )

    print("\nGenerated Text:")
    print(result[0]["generated_text"])
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch

# Load GPT model
model_name = "gpt2"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

print("GPT Chatbot Started (type 'exit' to stop)\n")

chat_history = ""

while True:

    user_input = input("You: ")

    if user_input.lower() == "exit":
        print("Chatbot: Goodbye!")
        break

    # Create conversation prompt
    prompt = chat_history + f"\nUser: {user_input}\nBot:"

    inputs = tokenizer.encode(prompt, return_tensors="pt")

    outputs = model.generate(
        inputs,
        max_new_tokens=50,
        do_sample=True,
        temperature=0.8,
        top_p=0.9,
        pad_token_id=tokenizer.eos_token_id
    )

    response = tokenizer.decode(outputs[0], skip_special_tokens=True)

    # Extract bot response
    bot_reply = response.split("Bot:")[-1].strip()

    print("Chatbot:", bot_reply)

    # Update history
    chat_history += f"\nUser: {user_input}\nBot: {bot_reply}"
```

```python
from transformers import BlenderbotTokenizer, BlenderbotForConditionalGeneration

# Load model
model_name = "facebook/blenderbot-400M-distill"

tokenizer = BlenderbotTokenizer.from_pretrained(model_name)
model = BlenderbotForConditionalGeneration.from_pretrained(model_name)

print("AI Chatbot Started (type 'exit' to stop)\n")

while True:

    user_input = input("You: ")

    if user_input.lower() == "exit":
        print("Chatbot: Goodbye!")
        break

    # Tokenize input
    inputs = tokenizer(user_input, return_tensors="pt")

    # Generate reply
    reply_ids = model.generate(
        **inputs,
        max_new_tokens=60,
        do_sample=True,
        temperature=0.7
    )

    response = tokenizer.decode(reply_ids[0], skip_special_tokens=True)

    print("Chatbot:", response)
```
